```python
import numpy as np
import matplotlib.pyplot as plt

# Your two points (endpoints of the diameter)
p1 = np.array([1, 2])
p2 = np.array([5, 6])

# 1. Calculate Center (Midpoint)
center = (p1 + p2) / 2

# 2. Calculate Radius (Half distance between points)
radius = np.linalg.norm(p1 - p2) / 2

# 3. Generate points for the circle
theta = np.linspace(0, 2*np.pi, 40)
x_circle = center[0] + radius * np.cos(theta)
y_circle = center[1] + radius * np.sin(theta)

# Plotting
plt.figure(figsize=(5,5))
plt.plot(x_circle, y_circle, label="Interpolated Circle")
plt.scatter([p1[0], p2[0]], [p1[1], p2[1]], color='red', label="Diameter Endpoints")
plt.gca().set_aspect('equal') # Crucial to make it look like a circle
plt.legend()
plt.show()
```

```python
p1 = np.array([1, 2])
p2 = np.array([5, 4])

# 1. Calculate Center and Radius
center = (p1 + p2) / 2
radius = np.linalg.norm(p1 - p2) / 2

# 2. Calculate the starting angle (angle of p1 relative to center)
start_angle = np.arctan2(p1[1] - center[1], p1[0] - center[0])

# 3. Generate angles for a semicircle (span of pi radians)
# Use start_angle to start_angle + np.pi for one side
# Use start_angle to start_angle - np.pi for the other side
theta = np.linspace(start_angle, start_angle + np.pi, 100)

# 4. Parametric equations for the arc
x_semicircle = center[0] + radius * np.cos(theta)
y_semicircle = center[1] + radius * np.sin(theta)

# Plotting
plt.figure(figsize=(6,6))
plt.plot(x_semicircle, y_semicircle, 'b-', label="Semicircle")
plt.scatter([p1[0], p2[0]], [p1[1], p2[1]], color='red', label="Diameter Points")
plt.gca().set_aspect('equal')
plt.legend()
plt.show()
```
