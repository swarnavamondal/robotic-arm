#### including libraries

```python
import numpy as np
import matplotlib.pyplot as plt  
```

```python
import seaborn as sns
sns.set_style("whitegrid")
```

#### array for ploting

```python
X = np.array([0, 9])
Y = np.array([0, 3])
```

#### ploting base graph

```python
plt.plot(X, Y, marker='o',linestyle='--',color = 'green',markerfacecolor='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Line Plot')
plt.show()
```

### linear function

```python
from scipy.interpolate import interp1d
lineareqn = interp1d(X, Y, kind='linear')
x2 = np.linspace(0, 9, num=10)
y2 = lineareqn(x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].plot(X, Y, marker='o',linestyle='--',color = 'green',markerfacecolor='red')
ax[0].set_title('Original Data Points')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[1].plot(x2, y2, marker='o',linestyle='--',color = 'blue',markerfacecolor='orange')
ax[1].set_title('Interpolated Data Points')
plt.show()
```

### quadratic function

```python
X= np.array([0,2,9])
Y= np.array([0,2,3])
quadraticeqn = interp1d(X, Y, kind='quadratic')
x2 = np.linspace(0, 9, num=100)
y2 = quadraticeqn(x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].plot(X, Y, marker='o',linestyle='--',color = 'green',markerfacecolor='red')
ax[0].set_title('Original Data Points')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[1].plot(x2, y2, marker='o',linestyle='--',color = 'blue',markerfacecolor='orange')
ax[1].set_title('Interpolated Data Points')
plt.show()
```

### cubic function

```python
X= np.array([0,2,4,9])
Y= np.array([0,2,2,3])
cubiceqn = interp1d(X, Y, kind='cubic')
x2 = np.linspace(0, 9, num=10)
y2 = cubiceqn(x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].plot(X, Y, marker='o',linestyle='--',color = 'green',markerfacecolor='red')
ax[0].set_title('Original Data Points')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[1].plot(x2, y2, marker='o',linestyle='--',color = 'blue',markerfacecolor='orange')
ax[1].set_title('Interpolated Data Points')
plt.show()
```

### slinear

```python
X= np.array([0,2,4,9])
Y= np.array([0,2,2,3])
slineareqn = interp1d(X, Y, kind='slinear')
x2 = np.linspace(0, 9, num=10)
y2 = slineareqn(x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].plot(X, Y, marker='o',linestyle='--',color = 'green',markerfacecolor='red')
ax[0].set_title('Original Data Points')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[1].plot(x2, y2, marker='o',linestyle='--',color = 'blue',markerfacecolor='orange')
ax[1].set_title('Interpolated Data Points')
plt.show()
```

```python
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x2, y2, marker='o',linestyle='--',color = 'blue',markerfacecolor='orange')
ax.set_title('Interpolated Data Points')
plt.show()
```
